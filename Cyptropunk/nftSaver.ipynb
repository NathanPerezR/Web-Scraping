```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os

def save_html(html,path):
  with open(path,'wb') as f:
      f.write(html)

def open_html(path):
  with open(path,'rb') as f:
      return f.read()
```

```python
MAX_TO_GRAB = 10 # this number must be between 0 and 10,000
```

```python
# saves the png's to the folder

# another 10,000 pages for the rest of the data
for i in range(MAX_TO_GRAB):
  punkOn = "{:04d}".format(i)

  # photo for each photo has predictable URL
  url = f'https://cryptopunks.app/public/images/cryptopunks/punk{punkOn}.png'
  r = requests.get(url)

  #save png
  local_output_path = "./output/png/" + punkOn + ".png"
  if os.path.isfile(local_output_path):
    print(f"{punkOn}: exist locally")
  else: 
    save_html(r.content, local_output_path)
    print(f"{punkOn}: saved locally")
```

```python
# saves the html files to a folder

# another 10,000 pages for the rest of the data
for i in range(MAX_TO_GRAB):
  punkOn = "{:04d}".format(i)

  # photo for each photo has predictable URL
  url = f'https://cryptopunks.app/cryptopunks/details/{punkOn}'
  r = requests.get(url)

  #save png
  local_output_path = "./output/html/" + punkOn + ".html"
  if os.path.isfile(local_output_path):
    print(f"{punkOn}: exist locally")
  else: 
    save_html(r.content, local_output_path)
    print(f"{punkOn}: saved locally")
```

```python


# another 10,000 pages for the rest of the data
for i in range(29):
  punkOn = "{:04d}".format(i)
  # formatted string for the file we are on
  pathToCurFile =  "./output/html/" + punkOn + ".html" 
  
  # no url needed since data is local
  html = open_html(pathToCurFile)
  soup = BeautifulSoup(html)

  soldGen = 0
  claim = 0
  detail = 0
  attributeGen = 0
  

  # Throw all data into a dataframe for export 
  soldGen       = (soup.select(".punk-history-row-sold a"))
  claim         = (soup.select_one(".punk-history-row-claim").text.strip())
  detail        = (soup.select_one(".row.detail-row a").text.strip())
  attributeGen  = (soup.select(".row.detail-row .row a"))
  claimed       = soup.select_one(".col-md-10.col-md-offset-1.col-xs-12 h4").text.strip()

  print(soldGen     ,    
        claim       ,
        detail      ,
        attributeGen,
        claimed     ,  )
```
